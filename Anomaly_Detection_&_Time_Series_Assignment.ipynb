Question 1: What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.

Answer 1 : Anomaly detection is the process of identifying rare data points, events, or patterns that deviate significantly from normal behavior within a dataset, often signaling issues like fraud, cyberattacks, or system malfunctions. Its three main types are point anomalies, which are individual, isolated data points that are statistically different (e.g., an unusually large purchase); contextual anomalies, which are data points that seem normal on their own but are anomalous in their specific context (e.g., high spending in winter); and collective anomalies, which are groups of data points that are unusual when considered together, even if individual points are not anomalous (e.g., a sudden, prolonged increase in network traffic).

Here are the three types of anomalies with examples:

1. Point Anomalies

Definition:

These are individual data points that are outliers or exceptions compared to the rest of the dataset.

Example:

A customer makes a single credit card transaction that is exceptionally large and far from their typical spending patterns. This single transaction is a point anomaly because its value is statistically unusual on its own.

2. Contextual Anomalies

Definition:

Data points that are considered anomalous only within a specific context or situation.

Example:

A significant rise in heating system energy consumption is normal during winter but would be a contextual anomaly during summer. The high consumption is normal when it's cold but out of place when it's hot.

3. Collective Anomalies

Definition:

A collection of related data points that, when viewed together, exhibit unusual behavior, even if no single point within the collection is itself an anomaly.

Example:

A sudden increase in login attempts from a particular IP address over an extended period, even if each individual attempt seems normal, could signal a collective anomaly indicative of a coordinated cyberattack.

Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

Answer 2 : Isolation Forest identifies outliers by isolating them using decision trees, making it fast and scalable for global outliers. DBSCAN, a clustering algorithm, finds anomalies as points that don't belong to any high-density cluster, excelling with irregular shapes and unknown cluster numbers but struggling with high-dimensional data. Local Outlier Factor (LOF) measures an outlier's degree of isolation based on its neighborhood's density, effective for local anomalies but less scalable and computationally intensive for high dimensions.

Here's a detailed comparison:

1. Isolation Forest

Approach:

This algorithm builds an ensemble of Isolation Trees (iTrees). It "isolates" outliers by randomly selecting a feature and then randomly selecting a split value for that feature to divide the data. Anomalies are easier to isolate, thus having shorter path lengths in the trees.

Suitable Use Cases:

Large, High-Dimensional Datasets: Its efficiency makes it ideal for identifying outliers in vast datasets.

Fraud Detection: Quickly detecting anomalous transactions in financial systems.
System Monitoring: Spotting unusual patterns in system faults or network intrusion.

Scalability: It is faster and more scalable than LOF.

Weakness:

Can struggle to detect local outliers, as it focuses more on global anomalies.

2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

Approach:

A clustering algorithm that groups together densely packed points (clusters) and marks sparse points as outliers or noise. It's a density-based method, requiring parameters for core distance and search distance to define clusters.

Suitable Use Cases:

Irregularly Shaped Clusters: It's excellent at discovering clusters of arbitrary shapes, unlike k-means.

Unknown Number of Clusters: It does not require the number of clusters to be specified beforehand.

Noise and Outlier Detection: Naturally handles noise by identifying points that don't belong to any density-based cluster.


Weakness:
Not ideal for high-dimensional data and can be sensitive to the choice of density parameters.
3. Local Outlier Factor (LOF)

Approach:

Measures the local density deviation of a given data point with respect to its neighbors. It calculates how isolated a point is compared to its local neighborhood, flagging points with significantly lower density than their neighbors as anomalies.

Suitable Use Cases:

Local Anomalies: Specifically designed to find anomalies that might not be apparent in the global view but are unusual within their local context.
Fraud Detection and Intrusion Detection: Identifying individual anomalies in complex systems.

Weakness:

Computationally expensive and does not scale well to high-dimensional data.

Question 3: What are the key components of a Time Series? Explain each with one example.

Answer 3 : The four key components of a time series are:-


1.*Trend* is the long-term direction of the data, either increasing or decreasing over time. For example, a company's annual revenue shows an increasing trend over several years

2.*Seasonal Variation* refers to regular, predictable fluctuations that repeat at fixed intervals within a year, like a monthly or quarterly pattern. For instance, sales of ice cream typically peak during the summer months every year.

3.*Cyclical Variation* includes long-term oscillations or waves that are not of a fixed period, often reflecting economic cycles. For example, periods of economic boom and recession, which repeat every few years, are a cyclical pattern.

4.*Irregular Variation* (also known as Noise) consists of unpredictable, random changes in the data caused by unforeseen events. A sudden drop in stock prices due to a natural disaster would be an example of an irregular variation.

Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

Answer 4 :
### **Definition of Stationarity**

A **stationary time series** is one whose **statistical properties do not change over time**. Specifically:

1. **Constant mean** → The average value of the series remains stable over time.
2. **Constant variance** → The spread (volatility) of the series does not increase or decrease with time.
3. **Constant autocovariance** → The relationship between observations (correlations) depends only on the lag between them, not on the actual time at which they are measured.

📌 Example: White noise is stationary; stock market prices usually are not (they trend).

---

### **Testing for Stationarity**

There are several methods to check if a series is stationary:

1. **Visual Inspection**

   * Plot the series. If it shows trends, seasonality, or changing variance, it’s likely non-stationary.

2. **Summary Statistics**

   * Split the data into different segments and compare means/variances. If they change significantly → non-stationary.

3. **Statistical Tests**

   * **Augmented Dickey-Fuller (ADF) Test**: Null hypothesis = non-stationary. If p-value < 0.05 → reject null → stationary.
   * **KPSS Test** (Kwiatkowski-Phillips-Schmidt-Shin): Null hypothesis = stationary. If p-value < 0.05 → reject null → non-stationary.
   * Using both ADF & KPSS together gives a more reliable conclusion.

---

### **Transforming Non-Stationary Series**

1. **Differencing**

   * Subtract the current observation from the previous one:

     $$
     Y'_t = Y_t - Y_{t-1}
     $$
   * Removes trend. If still non-stationary, take second differences.

2. **Log Transformation / Square Root**

   * Stabilizes variance (useful for series with exponential growth).

3. **Detrending**

   * Fit a trend (linear, polynomial, exponential) and subtract it.

4. **Seasonal Adjustment**

   * Remove seasonality by differencing with seasonal lag (e.g., $Y_t - Y_{t-12}$ for monthly data).

Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

Answer 5 :

## **1. AR (Autoregressive Model)**

* **Structure**:

  * Forecasts future values using a **linear combination of past values**.
  * AR(p):

    $$
    Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \dots + \phi_p Y_{t-p} + \epsilon_t
    $$
  * Here, $p$ = number of lags used.
* **Application**: Works well when a series shows correlation with its own past values (autocorrelation). Example: predicting stock returns.

---

## **2. MA (Moving Average Model)**

* **Structure**:

  * Uses **past forecast errors** (shocks/noise) to predict current values.
  * MA(q):

    $$
    Y_t = c + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \dots + \theta_q \epsilon_{t-q} + \epsilon_t
    $$
  * Here, $q$ = number of lagged errors used.
* **Application**: Useful when the data shows **short-term shock effects** (e.g., sudden demand surges).

---

## **3. ARIMA (Autoregressive Integrated Moving Average)**

* **Structure**:

  * Combines AR + differencing (I) + MA.
  * ARIMA(p, d, q):

    * **p** = AR terms
    * **d** = number of differences to make series stationary
    * **q** = MA terms
* **Application**: Suitable for **non-stationary series with trends**, after differencing. Example: forecasting sales over time.

---

## **4. SARIMA (Seasonal ARIMA)**

* **Structure**:

  * Extends ARIMA to handle **seasonality** explicitly.
  * SARIMA(p, d, q)(P, D, Q, s):

    * First part: ARIMA(p, d, q)
    * Second part (P, D, Q, s): Seasonal components

      * P = seasonal AR terms
      * D = seasonal differencing
      * Q = seasonal MA terms
      * s = seasonal period (e.g., 12 for monthly data with yearly cycle)
* **Application**: Useful when the data shows **repeated seasonal patterns** (e.g., electricity demand, monthly sales).

---

## **5. SARIMAX (SARIMA with Exogenous Variables)**

* **Structure**:

  * SARIMA + **external regressors (X)**.
  * Incorporates other independent variables that influence the target series.
  * Example equation:

    $$
    Y_t = \text{SARIMA terms} + \beta X_t + \epsilon_t
    $$
* **Application**: Best when the series depends not only on its own past values but also on **external drivers**.

  * Example: Forecasting demand using SARIMAX with marketing spend, promotions, or weather data as external regressors.

---

### **Quick Comparison Table**

| Model       | Key Idea                    | Handles Trend?            | Handles Seasonality? | Uses External Factors? | Example Use        |
| ----------- | --------------------------- | ------------------------- | -------------------- | ---------------------- | ------------------ |
| **AR**      | Past values                 | ❌ (needs stationary data) | ❌                    | ❌                      | Stock returns      |
| **MA**      | Past errors                 | ❌                         | ❌                    | ❌                      | Short-term shocks  |
| **ARIMA**   | AR + differencing + MA      | ✅                         | ❌                    | ❌                      | Sales trend        |
| **SARIMA**  | ARIMA + seasonal terms      | ✅                         | ✅                    | ❌                      | Energy demand      |
| **SARIMAX** | SARIMA + external variables | ✅                         | ✅                    | ✅                      | Sales + promotions |

Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components.



```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm

# Step 2: Load the AirPassengers dataset
# This dataset is available in statsmodels but needs to be loaded using get_rdataset
data = sm.datasets.get_rdataset("AirPassengers", "datasets").data

# Convert 'time' to datetime and set as index
data['time'] = pd.to_datetime(data['time'], origin='unix', unit='D')
data.set_index('time', inplace=True)

# Step 3: Plot the original time series
plt.figure(figsize=(12, 5))
plt.plot(data['value'], label='AirPassengers')
plt.title("Monthly AirPassengers Data")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()

# Step 4: Decompose the series
decomposition = seasonal_decompose(data['value'], model='multiplicative', period=12)

# Step 5: Plot decomposition results
fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.show()
```

Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Step 2: Load dataset
# Example: NYC Taxi Fare dataset has 'fare_amount' and 'trip_distance'
# For simplicity, let’s assume you have a CSV with these columns
# data = pd.read_csv("nyc_taxi_fare.csv")

# Simulated small dataset (replace this with actual NYC Taxi Fare data)
np.random.seed(42)
fare = np.random.normal(10, 2, 200)      # normal fares around $10
distance = np.random.normal(5, 1.5, 200) # normal distances around 5 miles

# Add some anomalies
fare = np.concatenate([fare, [50, 60, 70]])       # unusually high fares
distance = np.concatenate([distance, [30, 40, 50]])  # unusually long trips

data = pd.DataFrame({"fare_amount": fare, "trip_distance": distance})

# Step 3: Fit Isolation Forest
iso = IsolationForest(contamination=0.05, random_state=42)
data['anomaly'] = iso.fit_predict(data[['fare_amount', 'trip_distance']])

# Step 4: Separate anomalies and normal points
normal = data[data['anomaly'] == 1]
anomalies = data[data['anomaly'] == -1]

# Step 5: Visualize results
plt.figure(figsize=(10, 6))
plt.scatter(normal['trip_distance'], normal['fare_amount'],
            c='blue', label='Normal', alpha=0.6)
plt.scatter(anomalies['trip_distance'], anomalies['fare_amount'],
            c='red', label='Anomaly', marker='x', s=100)
plt.xlabel("Trip Distance (miles)")
plt.ylabel("Fare Amount ($)")
plt.title("NYC Taxi Fare Anomaly Detection using Isolation Forest")
plt.legend()
plt.show()
```

Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.

```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Step 2: Load the AirPassengers dataset
# This dataset is available in statsmodels but needs to be loaded using get_rdataset
data = sm.datasets.get_rdataset("AirPassengers", "datasets").data

# Convert 'time' to datetime and set as index
data['time'] = pd.to_datetime(data['time'], origin='unix', unit='D')
data.set_index('time', inplace=True)

# Step 3: Plot original series
plt.figure(figsize=(12, 5))
plt.plot(data['value'], label="AirPassengers")
plt.title("Monthly AirPassengers Data (1949-1960)")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()

# Step 4: Define and fit SARIMA model
# (p,d,q) = (1,1,1), seasonal (P,D,Q,s) = (1,1,1,12) for yearly seasonality
model = SARIMAX(data['value'],
                order=(1,1,1),
                seasonal_order=(1,1,1,12),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit(disp=False)

# Step 5: Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_ci = forecast.conf_int()

# Step 6: Plot results
plt.figure(figsize=(12,6))
plt.plot(data['value'], label="Observed")
plt.plot(forecast.predicted_mean, label="Forecast", color="red")

# Confidence intervals
plt.fill_between(forecast_ci.index,
                 forecast_ci.iloc[:, 0],
                 forecast_ci.iloc[:, 1], color="pink", alpha=0.3)

plt.title("SARIMA Forecast for AirPassengers (Next 12 Months)")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# 1. Create a synthetic numerical dataset
np.random.seed(42)
X_inliers = 0.3 * np.random.randn(100, 2)
X_inliers = np.r_[X_inliers + 2, X_inliers - 2] # Two clusters of inliers

X_outliers = np.random.uniform(low=-6, high=6, size=(20, 2)) # Random outliers
X = np.r_[X_inliers, X_outliers]

# 2. Apply Local Outlier Factor (LOF)
clf = LocalOutlierFactor(n_neighbors=20, contamination=0.1) # contamination is the expected proportion of outliers
y_pred = clf.fit_predict(X)
# The decision_function gives the opposite of the LOF score. Lower values indicate higher outlierness.
# We multiply by -1 to get actual LOF-like scores where higher is more outlier-like.
loff_scores = -clf.negative_outlier_factor_

# 3. Visualize the anomalies
plt.figure(figsize=(10, 8))
plt.title("Local Outlier Factor (LOF) Anomaly Detection")

# Plot inliers (predicted as non-outliers)
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], color='blue', s=50, label='Inliers')

# Plot outliers (predicted as outliers)
plt.scatter(X[y_pred == -1, 0], X[y_pred == -1, 1], color='red', s=100, marker='x', label='Outliers')

# Optionally, visualize the LOF scores as color intensity or size
# For better visualization, we can scale the LOF scores
# lof_scores_scaled = (loff_scores - loff_scores.min()) / (loff_scores.max() - loff_scores.min()) * 200 + 50
# plt.scatter(X[:, 0], X[:, 1], s=loff_scores_scaled, edgecolors='k', facecolors='none', alpha=0.6, label='LOF Score')


plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()

print("\n--- LOF Scores for the first 10 data points ---")
print(loff_scores[:10])
print("\n--- Predicted labels for the first 10 data points (1: Inlier, -1: Outlier) ---")
print(y_pred[:10])
```

Question 10: You are working as a data scientist for a power grid monitoring company. Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage.

 Explain your real-time data science workflow:

 ● How would you detect anomalies in this streaming data (Isolation Forest / LOF / DBSCAN)?

 ● Which time series model would you use for short-term forecasting (ARIMA / SARIMA / SARIMAX)?

 ● How would you validate and monitor the performance over time?

 ● How would this solution help business decisions or operations?


 Answer 10 :

 Real-Time Data Science Workflow for Energy Demand Forecasting & Anomaly Detection

1. Detecting Anomalies in Streaming Data

Preferred Algorithm: Isolation Forest (fast, scalable, robust).

Why not LOF/DBSCAN?

LOF → good for local density anomalies but heavier for streaming.

DBSCAN → clustering-based, not efficient in real-time.

Implementation: Train Isolation Forest on historical features (energy_usage, weather, time-of-day) → apply continuously to streaming batches.

2. Short-Term Forecasting Model

Model: SARIMAX

Captures trend + seasonality.

Allows external regressors (weather conditions, region info).

Forecast next few hours/days of demand (granularity = 15 min).

3. Validation & Monitoring

Offline Validation: Train-test split, use RMSE, MAE, MAPE.

Online Monitoring:

Rolling error metrics (last 24h).

Drift detection (ADWIN/DDM).

Retraining weekly/daily.

4. Business Impact

Grid Reliability → detect spikes → trigger backup supply.

Cost Optimization → efficient scheduling of generation.

Energy Trading → better buying/selling decisions.

Renewables Integration → handle variability in solar/wind

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX

# -----------------------------
# 1. Simulate Streaming Dataset
# -----------------------------
np.random.seed(42)
time_index = pd.date_range(start="2023-01-01", periods=24*30, freq="15min")  # 30 days of 15-min data
energy_usage = 200 + 20*np.sin(np.arange(len(time_index))/96*np.pi) + np.random.normal(0, 5, len(time_index))
weather_temp = 25 + 5*np.sin(np.arange(len(time_index))/144*np.pi) + np.random.normal(0, 1, len(time_index))

# Add anomalies at valid indices
energy_usage[200] = 400  # spike
energy_usage[500] = 50  # drop  (Changed from 1000 to 500)

df = pd.DataFrame({"timestamp": time_index,
                   "energy_usage": energy_usage,
                   "temperature": weather_temp})
df.set_index("timestamp", inplace=True)

# -----------------------------
# 2. Anomaly Detection (Isolation Forest)
# -----------------------------
iso = IsolationForest(contamination=0.01, random_state=42)
df["anomaly"] = iso.fit_predict(df[["energy_usage", "temperature"]])

# -----------------------------
# 3. Short-Term Forecasting (SARIMAX)
# -----------------------------
# Resample to hourly for smoother forecasting
df_hourly = df["energy_usage"].resample("H").mean()

# Train SARIMAX (using temperature as exogenous regressor)
model = SARIMAX(df_hourly,
                order=(1,1,1),
                seasonal_order=(1,1,1,24),  # daily cycle
                exog=df["temperature"].resample("H").mean(),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit(disp=False)

# Forecast next 24 hours
# Using the last 24 hourly temperature values from the simulated data as exogenous variables for forecasting
forecast_exog = df["temperature"].resample("H").mean().tail(24).values.reshape(-1, 1)
forecast = results.get_forecast(steps=24, exog=forecast_exog)
forecast_ci = forecast.conf_int()

# -----------------------------
# 4. Visualization
# -----------------------------
plt.figure(figsize=(14,6))

# Plot anomalies
plt.subplot(1,2,1)
plt.plot(df.index, df["energy_usage"], label="Energy Usage", alpha=0.7)
plt.scatter(df[df["anomaly"]==-1].index, df[df["anomaly"]==-1]["energy_usage"],
            color="red", label="Anomaly", marker="x", s=60)
plt.title("Anomaly Detection (Isolation Forest)")
plt.xlabel("Time")
plt.ylabel("Energy Usage (MW)")
plt.legend()

# Plot forecast
plt.subplot(1,2,2)
plt.plot(df_hourly.index, df_hourly, label="Observed")
plt.plot(forecast.predicted_mean.index, forecast.predicted_mean, label="Forecast", color="red")
plt.fill_between(forecast_ci.index,
                 forecast_ci.iloc[:,0],
                 forecast_ci.iloc[:,1], color="pink", alpha=0.3)
plt.title("Energy Demand Forecast (SARIMAX, Next 24h)")
plt.xlabel("Time")
plt.ylabel("Energy Usage (MW)")
plt.legend()

plt.tight_layout()
plt.show()
```
